```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load data
df = pd.read_csv(r"Downloads/cbb24.csv")
```

```python
df.head()
```

```python
### Bar Chart of Free Throw Rate (FTR) of Top 20 Teams
```

```python
# Selecting top 20 teams based on ranking
top_20_teams = df.sort_values(by='RK').head(20)

# Plot bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='TEAM', y='FTR', data=top_20_teams, palette='viridis')
plt.xticks(rotation=45)
plt.title('Free Throw Rate (FTR) of Top 20 Teams')
plt.xlabel('Team')
plt.ylabel('FTR')
plt.tight_layout()

plt.show()
```

```python
### Bar Chart of Average Wins Above Bubble (WAB) by Conference
```

```python
# Group data by conference and calculate the mean
conference_wab = df.groupby('CONF')['WAB'].mean().sort_values()

# Cplot bar chart of Wins Above Bubble (WAB) by conference
plt.figure(figsize=(12, 8))
sns.barplot(x=conference_wab.index, y=conference_wab.values, palette='viridis')
plt.xticks(rotation=90)
plt.xlabel('Conference')
plt.ylabel('Average Wins Above Bubble (WAB)')
plt.title('Average Wins Above Bubble (WAB) by Conference')
plt.tight_layout()

plt.show()
```

```python
### Average Effective Field Goal Percentage (EFG%) of Top 10 Teams
```

```python
# Selecting top 10 teams based on ranking
top_10_teams = df.sort_values(by='RK').head(10)

# Create a line chart for the EFG% of the top 10 teams
plt.figure(figsize=(12, 8))
sns.lineplot(x='TEAM', y='EFG%', data=top_10_teams, marker='o')
plt.xticks(rotation=45)
plt.title('Average Effective Field Goal Percentage (EFG%) of Top 10 Teams')
plt.xlabel('Team')
plt.ylabel('Effective Field Goal Percentage (EFG%)')
plt.grid(True)

plt.show()
```

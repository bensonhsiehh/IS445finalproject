```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load data
df = pd.read_csv(r"Downloads/cbb24.csv")
```

```python
df.head()
```

```python
### Box Plot of Key Features by Conference
```

```python
# Plot box plots
plt.figure(figsize=(20, 10))

# Subplot 1: Power Rating (BARTHAG) by Conference
plt.subplot(1, 3, 1)
sns.boxplot(x='CONF', y='BARTHAG', data=df)
plt.xticks(rotation=90)
plt.title('Power Rating (BARTHAG) by Conference')

# Subplot 2: Adjusted Tempo (ADJ_T) by Conference
plt.subplot(1, 3, 2)
sns.boxplot(x='CONF', y='ADJ_T', data=df)
plt.xticks(rotation=90)
plt.title('Adjusted Tempo (ADJ_T) by Conference')

# Subplot 3: Effective Field Goal Percentage (EFG%) by Conference
plt.subplot(1, 3, 3)
sns.boxplot(x='CONF', y='EFG%', data=df)
plt.xticks(rotation=90)
plt.title('Effective Field Goal Percentage (EFG%) by Conference')

plt.suptitle('Distribution of Key Metrics Across Conferences')
plt.tight_layout(rect=[0, 0.03, 1, 0.95])

plt.show()
```

```python
### Stacked Bar Chart of Shooting Percentages for Top 20 Teams
```

```python
# Selecting top 20 teams based on ranking
top_20_teams = df.sort_values(by='RK').head(20)

# Plot stacked bar chart
plt.figure(figsize=(12, 8))

# Plot the two-point shooting percentages (2P_O)
plt.bar(top_20_teams['TEAM'], top_20_teams['2P_O'], label='2P_O', color='royalblue')

# Plot the three-point shooting percentages (3P_O) on top of the two-point percentages (2P_O)
plt.bar(top_20_teams['TEAM'], top_20_teams['3P_O'], bottom=top_20_teams['2P_O'], label='3P_O', color='lightblue')

plt.xticks(rotation=90)
plt.ylabel('Percentage')
plt.xlabel('Team')
plt.title('Composition of Two-Point and Three-Point Shooting Percentages for Top 20 Teams')
plt.legend()
plt.tight_layout()

plt.show()
```

```python
### Histogram of the Distribution for the Adjusted Offensive Efficiently
```

```python
# plot histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['ADJOE'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Adjusted Offensive Efficiency (ADJOE)')
plt.xlabel('Adjusted Offensive Efficiency (ADJOE)')
plt.ylabel('Number of Teams')
plt.grid(True)

```
